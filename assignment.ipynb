```python
pew = pew.rename(columns={
    '<$10k':'less_than_10k',
    '$10-20k' : '10_to_20k',
    '$20-30k' : '20_to_30k',
    '$30-40k' : '30_to_40k',
    '$40-50k' : '40_to_50k',
    '$50-75k' : '50_to_75k'
})

pew
```

```python
pew = pd.melt(pew, id_vars=['religion'], value_vars=['less_than_10k', '10_to_20k','20_to_30k', '30_to_40k', '40_to_50k', '50_to_75k'],
             value_name = 'frequency', var_name='income_range')

pew
```
